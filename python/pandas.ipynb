```python
# import libraries
import re
import numpy as np
import pandas as pd
```

```python
# import data
df1 = pd.read_csv("../data/iris/iris_sample1.csv")
df2 = pd.read_csv("../data/iris/iris_sample2.csv")
```

```python
# concatnate df1 and df2 rows
df = pd.concat(objs=[df1, df2], ignore_index=True)
df
```

```python
# Add a list as column
test_col = ["test1", "test2", "test3", "test4", "test5", "test6", "test7", "test8", "test9", "test10"]
df["TestColumn"] = test_col
df
```

```python
# Drop a column
df = df.drop(["TestColumn"], axis=1)
df
```

```python
# Drop duplicate rows
df.drop_duplicates(["Species","PetalLengthCm"])
```

```python
# remove null entries
df = df[df["Species"].notnull()]
df
```

```python
# create dictionary with one column as key and one as value
dict_id_date = dict(zip(df.Id,df.Species))
dict_id_date
```

```python
# Keep rows meetig a condition using apply and a function
def check_condition(text):
    if re.search("setosa", text):
        return False
    else:
        return True

df.insert(loc=6, column='Condition', value=df.apply(lambda x: check_condition(str(x[5])), axis=1))
df.loc[df["Condition"]]
```

```python
# Keep rows equal to condition
df[df["Species"]=="Iris-setosa"]
```

```python
# Keep rows if column value in list and not in list
check_list = [1,2]
df[df.Id.isin(check_list)]
```

```python
# Keep rows if not in list
check_list = [1,2]
df[~df.Id.isin(check_list)]
```

```python
# Keep rows matching a regex or pattern
df[df["Species"].str.match(".*setosa")]
```

```python
# Kepp rows if column contains a regex
df[df["Species"].str.contains("versicolor")]
```

```python
# Create a new dataframe from lists
col1 = [1,2,3,4,5]
col2 = ["a","b","c","d","e"]
pd.DataFrame({"col1":col1, "col2":col2})
```

```python
# Put each element of a list in a column to separate rows
col = [["a","b","c"],["d","e"],"f"]
pd.DataFrame({"col":col}).apply(lambda x: pd.Series(x['col']),axis=1).stack().reset_index(level=1, drop=True)
```
